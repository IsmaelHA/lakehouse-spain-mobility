```python
SQL("""CREATE TABLE silver_mobility_trips AS
SELECT
    CAST(date AS DATE) AS date,
    CAST(hour_period AS INTEGER) AS hour,
    origin_zone,
    destination_zone,
    distance_category,
    origin_activity,
    destination_activity,
    origin_study_possible IN ('si','sí') AS origin_study_possible,
    destination_study_possible IN ('si','sí') AS destination_study_possible,
    CAST(residence_province AS INTEGER) AS residence_province,
    income_bracket,
    age_bracket,
    sex,
    CAST(trips AS DOUBLE) AS trips,
    CAST(trips_km AS DOUBLE) AS trips_km,
    
    -- derived field
    CASE 
        WHEN EXTRACT(dow FROM CAST(date AS DATE)) IN (6,0) THEN 'weekend'
        ELSE 'weekday'
    END AS day_type

FROM bronze_raw_mobility_trips;
""")
```
